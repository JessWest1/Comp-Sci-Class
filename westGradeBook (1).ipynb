```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(columns = ['Student'])
df
```

```python
def alphaB(strng):
    global df
    if(len(df)>=1):
        dfNew = pd.DataFrame(columns = df.keys())
        dfNew.loc[0] = " "
        dfNew.loc[0]['Student'] = strng 
        df = pd.concat([df, dfNew])
        df.reset_index(drop = True, inplace = True)
    else:
        df.loc[0] = " "
        df.loc[0]['Student'] = strng
        
                
            
            
```

```python
def updateTest(strng):
    nameS = input("Enter the name of the student whos score you wish to add:")
    score = input("Enter the score of the test: ")
    indx = df[df['Student'] == nameS].index
    df.at[indx, strng] = int(score) # set_value doesnt work anymore
```

```python
def removeT(strng):
    df.drop([strng], axis = 1, inplace = True)   
```

```python
def removeS(strng):
    indx = df[df['Student'] == strng].index
    df.drop(indx, inplace = True)
```

```python
#add test column

def newTest(strng):
    df[strng] = " "
```
